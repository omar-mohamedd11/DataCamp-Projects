```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

Filtering movies only from the 90s

```python
movies_from_90s = netflix_df[(netflix_df['release_year'] >= 1990) &
                    (netflix_df['release_year'] <= 1999)]
```

Visualizing the most frequent movie duration during the 90s

```python
plt.hist(movies_from_90s['duration'],bins = 15)
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of movies')
plt.show()
duration = 100
```

Counting the action short movies from the 90s

```python
action_90s_df = movies_from_90s[movies_from_90s['genre'] == 'Action']
short_movie_count = 0
for lab,row in action_90s_df.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1
        
print('There is ' + str(short_movie_count) + ' action short 90s movies ')
```
